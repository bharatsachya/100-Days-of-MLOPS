```python
import requests
response = requests.get("https://www.eastoftheweb.com/short-stories/UBooks/Verd.shtml").content
response
```

```python
from bs4 import BeautifulSoup
bs4 = BeautifulSoup(response, "html.parser")
bs4.prettify()
```

```python
book = bs4.find_all("p")
```

```python
from re import I
var = ""
for i in book:
  var += i.text
print(var)
```

```python
text = "".join(i.text for i in book).strip("\n")
```

```python
import re
result = re.split(r'[\s,"/:?)(]',text)
print(result)
```

```python
result = [item.strip() for item in result if item.strip()]
print(result)
```

```python
list = sorted(set(result))
```

```python
vocab = {item:index for index,item in enumerate(result)}
```

```python
vocab
```

```python
class simpleTokenizerV1:
  def __init__(self, vocab):
    self.str_to_int = vocab
    # Create the reverse mapping from index to string
    self.int_to_str = {index: item for item, index in vocab.items()}

  def encode(self, text):
    preprocessed = text.lower()
    tokens = preprocessed.split()
    # Handle tokens not in the vocabulary
    return [self.str_to_int.get(token, -1) for token in tokens]

  def decode(self, list):
    # Map integers back to strings and join them with spaces
    text = " ".join([self.int_to_str.get(i, "[UNK]") for i in list])
    return text
```

```python
tokenizer = simpleTokenizerV1(vocab)
text = "rich widow and"
list = tokenizer.encode(text)
print(list)
orginal_text = tokenizer.decode(list)
print(orginal_text)
```
